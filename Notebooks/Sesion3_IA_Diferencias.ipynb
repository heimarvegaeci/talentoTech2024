# Bloque 1: Fundamentos de la Inteligencia Artificial

## Sesión 3: Diferencias entre IA, ML y DL

### Teoría

#### Diferencias Fundamentales entre Inteligencia Artificial (IA), Machine Learning (ML) y Deep Learning (DL)

**Inteligencia Artificial (IA):**
- **Definición:** La Inteligencia Artificial es un campo amplio de la informática que se centra en crear sistemas capaces de realizar tareas que normalmente requieren inteligencia humana. Estas tareas incluyen el reconocimiento de voz, toma de decisiones, traducción de idiomas, y mucho más.
- **Enfoque General:** La IA abarca cualquier técnica que permita a las máquinas imitar el comportamiento humano inteligente. Es un campo que incluye muchas subdisciplinas, como el aprendizaje automático, el procesamiento del lenguaje natural, la visión por computadora, entre otras.
- **Tipos de IA:** Se divide en IA débil (Narrow AI), que está diseñada para realizar una tarea específica, e IA fuerte (General AI), que tiene la capacidad de realizar cualquier tarea cognitiva humana.
- **Ejemplos de Aplicación:**
  - **Asistentes Virtuales:** Como Siri o Alexa, que comprenden y responden a comandos de voz.
  - **Automatización de Procesos:** Robots que realizan tareas repetitivas en fábricas.

**Machine Learning (ML):**
- **Definición:** El Machine Learning es una rama de la IA que permite a las máquinas aprender de datos, identificar patrones y tomar decisiones con mínima intervención humana. En lugar de ser explícitamente programadas para realizar una tarea, las máquinas aprenden de ejemplos y ajustan sus comportamientos en consecuencia.
- **Enfoque en Datos:** Los algoritmos de ML se entrenan en grandes volúmenes de datos. A través del proceso de entrenamiento, los algoritmos mejoran su precisión y efectividad.
- **Tipos de Aprendizaje:**
  - **Supervisado:** Se entrena con un conjunto de datos etiquetado donde el modelo aprende a partir de ejemplos previos.
  - **No Supervisado:** Se trabaja con datos no etiquetados, donde el modelo busca patrones o agrupaciones por sí mismo.
  - **Aprendizaje por Refuerzo:** El modelo aprende a tomar decisiones mediante la obtención de recompensas o penalizaciones por las acciones que realiza.
- **Ejemplos de Aplicación:**
  - **Sistemas de Recomendación:** Como los que usan Netflix o Amazon para sugerir productos o películas.
  - **Detección de Fraude:** Algoritmos que analizan patrones en transacciones para identificar actividades sospechosas.

**Deep Learning (DL):**
- **Definición:** El Deep Learning es una subcategoría del Machine Learning que utiliza redes neuronales artificiales con múltiples capas (deep neural networks) para modelar patrones complejos en grandes volúmenes de datos. Estas redes son capaces de aprender representaciones jerárquicas de los datos, lo que les permite realizar tareas como la clasificación de imágenes, el reconocimiento de voz, y más.
- **Redes Neuronales:** Las redes neuronales profundas están inspiradas en la estructura del cerebro humano y consisten en múltiples capas de neuronas artificiales que transforman la entrada en una salida a través de operaciones matemáticas.
- **Entrenamiento:** Requiere grandes cantidades de datos y una gran potencia computacional. Se entrenan mediante algoritmos de optimización, como la retropropagación, que ajusta los pesos de las conexiones neuronales para minimizar el error.
- **Ejemplos de Aplicación:**
  - **Reconocimiento de Imágenes:** Usado en aplicaciones de visión por computadora para identificar objetos en imágenes y videos.
  - **Procesamiento del Lenguaje Natural (PLN):** Para la traducción automática, chatbots avanzados, y análisis de sentimientos.

### Ejemplo

**Comparativa entre un Sistema Basado en Reglas y un Sistema de Machine Learning para Clasificación de Imágenes**

**Sistema Basado en Reglas:**
- **Funcionamiento:** Un sistema basado en reglas utiliza un conjunto de reglas definidas por expertos para tomar decisiones o realizar clasificaciones. En el caso de la clasificación de imágenes, podría usar reglas como "si el objeto tiene cuatro patas y un rabo, entonces es un perro".
- **Ventajas:**
  - **Simplicidad:** Fácil de entender y programar para problemas simples.
  - **Transparencia:** Las decisiones son completamente rastreables y explicables, ya que se basan en reglas explícitas.
- **Desventajas:**
  - **Rigidez:** No se adapta bien a situaciones nuevas o imprevistas. Si una imagen no coincide exactamente con las reglas predefinidas, el sistema fallará.
  - **Escalabilidad:** Dificultad para manejar tareas complejas con muchas variables o grandes volúmenes de datos.
- **Aplicación Típica:** Reconocimiento básico de patrones o procesos de toma de decisiones simples, como los sistemas de control de acceso.

**Sistema de Machine Learning:**
- **Funcionamiento:** Un sistema de ML aprende de ejemplos etiquetados durante una fase de entrenamiento. En la clasificación de imágenes, el sistema se entrena con miles de imágenes etiquetadas como "perro" o "gato", y aprende a reconocer patrones característicos de cada categoría.
- **Ventajas:**
  - **Adaptabilidad:** Puede mejorar su precisión con el tiempo al ser expuesto a más datos. Es capaz de manejar situaciones nuevas que no se habían considerado inicialmente.
  - **Escalabilidad:** Puede manejar grandes volúmenes de datos y complejidades, permitiendo la clasificación de imágenes con alta precisión.
- **Desventajas:**
  - **Opacidad:** Las decisiones de un modelo de ML pueden ser difíciles de interpretar, ya que no se basan en reglas explícitas.
  - **Requiere Datos y Computación:** Necesita grandes cantidades de datos para entrenarse correctamente, y recursos computacionales significativos.
- **Aplicación Típica:** Clasificación de imágenes, reconocimiento facial, sistemas de diagnóstico médico basados en imágenes.

### Comparativa

- **Sistema Basado en Reglas:**
  - Rígido y no escalable, adecuado para problemas simples y transparentes.
- **Sistema de ML:**
  - Flexible y escalable, capaz de manejar grandes volúmenes de datos, pero puede carecer de transparencia.

### Actividad

**En Pequeños Grupos, Crear un Cuadro Comparativo que Resuma las Diferencias entre IA, ML y DL**

**Objetivo de la Actividad:**
- **Comprender Diferencias Clave:** Los estudiantes identificarán y organizarán las diferencias clave entre IA, ML y DL en un formato visual que facilite la comprensión y comparación.

**Instrucciones:**
- **Formación de Grupos:** Los estudiantes se organizarán en grupos de 3-4 personas.
- **Investigación:** Cada grupo revisará los conceptos estudiados en clase sobre IA, ML y DL, incluyendo definiciones, ejemplos, aplicaciones, ventajas, y desventajas.
- **Desarrollo del Cuadro Comparativo:** Utilizando herramientas digitales como Google Sheets, Excel, o incluso en papel, los grupos crearán un cuadro comparativo que incluya:
  - **Definición:** ¿Qué es IA, ML, y DL?
  - **Ámbito de Aplicación:** ¿En qué tipos de problemas se utiliza cada uno?
  - **Ventajas y Desventajas:** ¿Cuáles son los puntos fuertes y débiles de cada enfoque?
  - **Ejemplos:** Un ejemplo específico para cada categoría.
- **Presentación:** Los grupos presentarán sus cuadros comparativos al resto de la clase, explicando las decisiones que tomaron al organizar la información.

**Criterios de Evaluación:**
- **Exactitud y Comprensión:** Se evaluará la precisión de la información presentada y la comprensión de los conceptos por parte de los estudiantes.
- **Claridad y Organización:** El cuadro comparativo debe ser claro, bien organizado y fácil de entender.
- **Participación en Grupo:** Evaluación del trabajo en equipo y la contribución de cada miembro al resultado final.
