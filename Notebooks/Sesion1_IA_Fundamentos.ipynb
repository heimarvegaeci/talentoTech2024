# Bloque 1: Fundamentos de la Inteligencia Artificial

## Sesión 1: Introducción a la IA

### Teoría

**Concepto de Inteligencia Artificial (IA):**
La Inteligencia Artificial (IA) se refiere a la simulación de procesos de inteligencia humana por parte de máquinas, especialmente sistemas informáticos. Estas capacidades incluyen el aprendizaje (adquisición de información y reglas para el uso de la información), el razonamiento (usando las reglas para llegar a conclusiones aproximadas o definitivas), y la autocorrección.

**Evolución Histórica de la IA:**
1. **1940-1950: Orígenes y Primeros Conceptos:**
   - **Turing y la Máquina Universal (1936):** Alan Turing propuso la idea de una "máquina universal" capaz de realizar cualquier cálculo matemático. Su trabajo sentó las bases de la informática moderna.
   - **La Prueba de Turing (1950):** Turing sugirió un método para evaluar la inteligencia de una máquina, conocido como la Prueba de Turing.

2. **1956: La Conferencia de Dartmouth:**
   - **Nacimiento Formal de la IA:** En esta conferencia, John McCarthy acuñó el término "Inteligencia Artificial". La IA fue reconocida como un campo de estudio académico.

3. **1960-1970: Primeras Aplicaciones y Sistemas Expertos:**
   - **Sistemas Basados en Reglas:** Primeros sistemas que utilizaban reglas predefinidas para tomar decisiones. Ejemplo: Eliza, un programa que simulaba una conversación terapéutica.
   - **Sistemas Expertos (Años 70):** Programas que emulan la toma de decisiones humanas en campos específicos, como la medicina (MYCIN) y la geología (DENDRAL).

4. **1980-1990: Invierno de la IA y Resurgimiento:**
   - **Invierno de la IA:** Una época de estancamiento debido a expectativas no cumplidas y limitaciones tecnológicas.
   - **Resurgimiento:** Con el aumento de la potencia de procesamiento y la disponibilidad de grandes volúmenes de datos, la IA comenzó a avanzar nuevamente. El uso de redes neuronales se popularizó.

5. **2000-Presente: Revolución del Aprendizaje Automático y DL:**
   - **Aprendizaje Automático (ML):** Desarrollo de algoritmos que permiten a las máquinas aprender de los datos sin ser programadas explícitamente para cada tarea.
   - **Deep Learning (DL):** Una subdisciplina de ML que se enfoca en redes neuronales profundas, que imitan el funcionamiento del cerebro humano y son capaces de procesar grandes cantidades de datos no estructurados.

**Diferencias entre IA, Machine Learning (ML) y Deep Learning (DL):**
- **IA (Inteligencia Artificial):** Campo amplio que engloba cualquier técnica que permita a las máquinas emular la inteligencia humana.
- **ML (Machine Learning):** Subcampo de la IA que se enfoca en el desarrollo de algoritmos que permiten a las máquinas aprender de los datos y hacer predicciones o decisiones basadas en esos datos.
- **DL (Deep Learning):** Subcampo de ML que utiliza redes neuronales profundas para aprender y tomar decisiones. Es especialmente útil en tareas complejas como el reconocimiento de imágenes y el procesamiento del lenguaje natural.

### Ejemplo

**Chatbots: Evolución desde Sistemas de Respuesta Fija a Sistemas Sofisticados de IA:**

1. **Chatbots Básicos (Eliza, 1966):**
   - **Funcionamiento:** Eliza era un programa de computadora diseñado para imitar la interacción de un psicoterapeuta con el paciente, respondiendo basándose en reglas predefinidas de coincidencia de patrones de texto. Este chatbot no entendía realmente el lenguaje, sino que simplemente seguía un conjunto de reglas para dar respuestas que parecieran coherentes.

2. **Chatbots Basados en ML:**
   - **Funcionamiento:** Con el avance de ML, los chatbots comenzaron a aprender de grandes conjuntos de datos de conversaciones. Un ejemplo sería el uso de modelos de clasificación que seleccionan la mejor respuesta basada en la entrada del usuario.
   - **Aplicación:** Estos chatbots se utilizan en atención al cliente, proporcionando respuestas automáticas a consultas comunes y aprendiendo de las interacciones pasadas para mejorar con el tiempo.

3. **Chatbots de IA Avanzada (GPT, BERT):**
   - **Funcionamiento:** Utilizan modelos de DL que procesan el lenguaje natural, como GPT (Generative Pre-trained Transformer) de OpenAI o BERT (Bidirectional Encoder Representations from Transformers) de Google. Estos modelos no solo pueden entender el contexto, sino también generar respuestas humanas y participar en conversaciones complejas.
   - **Aplicación:** Estos chatbots son capaces de realizar tareas más sofisticadas, como generar contenido personalizado, traducir idiomas en tiempo real, y comprender la intención del usuario para ofrecer recomendaciones más precisas.

### Actividad

**Investiga tres aplicaciones de IA en tu vida diaria y discútelas en clase:**

1. **Objetivo de la Actividad:**
   - **Comprender la Presencia de la IA:** Los estudiantes investigarán cómo la IA afecta diferentes aspectos de su vida diaria, reconociendo su omnipresencia en tecnologías comunes.

2. **Instrucciones:**
   - **Investigación Individual:** Cada estudiante identificará tres aplicaciones de IA que usan en su vida diaria. Ejemplos pueden incluir asistentes virtuales como Siri o Alexa, recomendaciones de películas en Netflix, o sistemas de navegación como Google Maps.
   - **Discusión en Clase:** En la siguiente sesión, los estudiantes presentarán sus hallazgos y discutirán cómo estas aplicaciones utilizan IA, ML, o DL, y cómo han mejorado su experiencia diaria.

3. **Criterios de Evaluación:**
   - **Profundidad de la Investigación:** Se evaluará si el estudiante ha identificado correctamente el uso de la IA en las aplicaciones seleccionadas y si ha comprendido cómo funcionan.
   - **Participación en la Discusión:** Se evaluará la capacidad del estudiante para discutir sus hallazgos y relacionar sus ejemplos con los conceptos teóricos aprendidos.
