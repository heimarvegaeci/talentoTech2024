# Bloque 1: Fundamentos de la Inteligencia Artificial

## Sesión 2: Evolución y Categorías de la IA

### Teoría

**Evolución de la IA: IA Débil vs. IA Fuerte**

1. **IA Débil (Weak AI):**
- **Concepto:** También conocida como IA estrecha, se refiere a sistemas diseñados y entrenados para realizar tareas específicas sin tener conciencia ni entendimiento real. Estos sistemas imitan algunas capacidades humanas, pero no poseen un conocimiento general ni la capacidad de razonar fuera de su dominio.
- **Características:**
  - **Especialización:** Están diseñados para resolver problemas específicos, como jugar al ajedrez, conducir un vehículo o reconocer rostros.
  - **No Conscientes:** Carecen de auto-conciencia o entendimiento más allá de la tarea para la cual fueron programados.
- **Ejemplos:** Siri de Apple, asistentes de voz como Alexa, y sistemas de recomendación de plataformas como Netflix.

2. **IA Fuerte (Strong AI):**
- **Concepto:** Se refiere a sistemas con capacidades cognitivas equivalentes a las de un ser humano, con la capacidad de razonar, planificar, aprender y tener conciencia de sí mismos. Este tipo de IA todavía es teórico y no ha sido desarrollado en la práctica.
- **Características:**
  - **Generalización:** Capaz de realizar cualquier tarea cognitiva que un humano pueda realizar, sin necesidad de programación específica para cada tarea.
  - **Consciencia:** Se postula que podrían tener conciencia y sentimientos, aunque este es un tema de debate filosófico.
- **Ejemplos:** A día de hoy, la IA fuerte sigue siendo un concepto hipotético, representado en la ciencia ficción por personajes como HAL 9000 en '2001: Odisea del Espacio' o Data en 'Star Trek'.

**Categorías de IA: Sistemas Expertos y Sistemas Basados en Reglas**

1. **Sistemas Expertos:**
- **Definición:** Un sistema experto es un programa de computadora que emula la toma de decisiones de un ser humano experto en un dominio específico. Se utilizan para resolver problemas complejos dentro de un campo particular, basándose en una base de conocimiento y reglas de inferencia.
- **Componentes:**
  - **Base de Conocimiento:** Conjunto de hechos y reglas sobre un área específica de conocimiento.
  - **Motor de Inferencia:** Algoritmo que aplica las reglas a los hechos conocidos para deducir nuevas informaciones o decisiones.
  - **Interfaz de Usuario:** Medio a través del cual los usuarios interactúan con el sistema, proporcionando datos y recibiendo conclusiones o recomendaciones.
- **Ejemplos:**
  - **MYCIN:** Uno de los primeros sistemas expertos, desarrollado en los años 70 para diagnosticar infecciones bacterianas y recomendar tratamientos.
  - **DENDRAL:** Sistema desarrollado para identificar estructuras moleculares complejas en química.

2. **Sistemas Basados en Reglas:**
- **Definición:** Son sistemas que operan mediante el uso de reglas 'si-entonces' para tomar decisiones o inferir nuevas informaciones. Estos sistemas son más simples que los sistemas expertos y no necesariamente requieren una base de conocimiento extensa.
- **Características:**
  - **Reglas Predefinidas:** El sistema sigue un conjunto de reglas definidas por programadores humanos.
  - **Aplicaciones Simples:** Se utilizan en tareas que no requieren un razonamiento profundo, como el procesamiento de transacciones o control de procesos industriales.
- **Ejemplos:** Un sistema de control de acceso que permite o niega la entrada basado en reglas predefinidas.

### Ejemplo

**Análisis de cómo los Sistemas Expertos han sido utilizados en Diagnósticos Médicos**

1. **Sistemas Expertos en Medicina:**
- **Contexto:** En el ámbito de la medicina, los sistemas expertos se han utilizado para apoyar a los médicos en el diagnóstico de enfermedades y la recomendación de tratamientos. Estos sistemas no reemplazan al médico, sino que proporcionan una segunda opinión basada en un vasto conjunto de datos médicos.

2. **Caso de Estudio: MYCIN:**
- **Función:** MYCIN fue un sistema experto desarrollado en la Universidad de Stanford en la década de 1970 para ayudar en el diagnóstico de infecciones bacterianas y recomendar antibióticos adecuados, incluyendo la dosis.
- **Base de Conocimiento:** Contenía una base de conocimiento con información sobre bacterias, infecciones y antibióticos, junto con reglas para asociar síntomas con posibles infecciones.
- **Motor de Inferencia:** Utilizaba reglas 'si-entonces' para deducir cuál era la infección probable y recomendar el tratamiento adecuado.
- **Impacto:** Aunque MYCIN nunca se implementó clínicamente debido a cuestiones legales y éticas, fue pionero en el uso de la IA en la medicina y sentó las bases para futuros sistemas expertos en diagnóstico.

3. **Evolución hacia la IA Actual:**
- **Integración con IA y ML:** Hoy en día, los sistemas expertos se han integrado con técnicas de aprendizaje automático para mejorar la precisión y adaptar las recomendaciones basadas en datos actuales. Los sistemas de diagnóstico modernos utilizan grandes volúmenes de datos, redes neuronales y aprendizaje profundo para identificar patrones y ofrecer diagnósticos más precisos y personalizados.

**Crear una Línea de Tiempo Interactiva con los Hitos más Importantes en la Evolución de la IA**

1. **Objetivo de la Actividad:**
- **Entender la Historia de la IA:** Los estudiantes desarrollarán una comprensión profunda de la evolución de la IA, identificando y ordenando cronológicamente los eventos y avances clave en el campo.

2. **Instrucciones:**
- **Investigación:** Cada estudiante o grupo investigará un período específico en la historia de la IA, identificando hitos importantes como la creación de la Máquina de Turing, la Conferencia de Dartmouth, el desarrollo de sistemas expertos como MYCIN, y avances modernos como el surgimiento del deep learning.
- **Herramientas:** Utilizarán herramientas digitales como [TimelineJS](https://timeline.knightlab.com/) para crear una línea de tiempo interactiva. Deberán incluir descripciones detalladas, imágenes, y enlaces a recursos adicionales para cada hito.
- **Presentación:** Una vez completada, cada grupo presentará su sección de la línea de tiempo, explicando la importancia de los hitos y cómo contribuyeron al desarrollo de la IA.

3. **Criterios de Evaluación:**
- **Exactitud y Relevancia:** Se evaluará la precisión de la información y la relevancia de los hitos seleccionados.
- **Creatividad y Presentación:** La creatividad en la presentación de la línea de tiempo y la claridad en la explicación de los eventos.
- **Colaboración en Grupo:** Se evaluará cómo trabajaron en equipo, su capacidad de investigación y la integración de su trabajo en la línea de tiempo general.
